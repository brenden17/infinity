```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def func(x, a, b):
    return a * x + b

x = np.linspace(0, 10, 100)
y = func(x, 1, 2)

yn = y + 0.9 * np.random.normal(size=len(x))

popt, pcov = curve_fit(func, x, yn)

#fig = plt.figure()
#ax = fig.add_subplot(111)

plt.subplot(211)
plt.plot(yn, 'b-', func(x, popt[0], popt[1]), 'g-')
plt.xlabel('time')
plt.ylabel('s1 and s2')
plt.grid(True)

def func2(x, a, b, c):
    return a*np.exp(-(x-b)**2/(2*c**2))

y = func2(x, 1, 2, 3)

yn = y + 0.1 * np.random.normal(size=len(x))

popt, pcov = curve_fit(func2, x, yn)
print popt

plt.subplot(212)
plt.plot(yn, 'ob', func2(x, popt[0], popt[1], popt[2]), 'g-')
plt.xlabel('time')
plt.ylabel('s1 and s2')
plt.grid(True)
```

```python
import numpy as np
from scipy.cluster import vq
import matplotlib.pyplot as plt

c1 = np.random.randn(100, 2) + 10

c2 = np.random.rand(30, 2) + 3
c3 = np.random.randn(20, 2)
data = np.vstack([c1, c2, c3])

centroids, variance = vq.kmeans(data, 3)

identified, distance = vq.vq(data, centroids)

#plt.subplot(211)
plt.subplot(211)
plt.plot(data[:, 0], data[:, 1],  'bo', centroids[:, 0], centroids[:, 1], 'ro')
plt.xlabel('time')
plt.ylabel('s1 and s2')
plt.grid(True)

plt.subplot(212)
plt.plot(data[identified==0][:,0], data[identified==0][:,1], 'bo', data[identified==1][:,0], data[identified==1][:,1], 'ro',  data[identified==2][:,0], data[identified==2][:,1], 'go')
plt.xlabel('time')
plt.ylabel('s1 and s2')
plt.grid(True)

```
