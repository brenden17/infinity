```python
"""
Talylor series for e
"""
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import factorial

x = np.linspace(-2, 2, 1000)

def f(x):
    return x**2 + 2*x +3
def f1(x):
    return x*0 +6
def f2(x):
    return 4*(x-1)
def f3(x):
    return (x-1)**2

plt.figsize(5,5)
plt.plot(x, f(x), '-')
plt.plot(x, f1(x), '-')
plt.plot(x, f2(x), '-')
plt.plot(x, f3(x), '-')
plt.plot(x, f1(x) + f2(x) + f3(x), 'b')

plt.axhline(y=0)
plt.grid()
```
